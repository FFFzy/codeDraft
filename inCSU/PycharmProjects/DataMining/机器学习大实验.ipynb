```python
def plot_line_bar(data, y1_max, y1_range, y2_max, y2_range):  #双轴组合图
    x = data.index
    y1 = data['count']
    y2 = data['播放量']
    f, ax1 = plt.subplots(figsize=(10, 5))
    ax1.bar(x, y1, color='#348498', label='发布视频数')
    ax2 = ax1.twinx()  # 双轴图关键
    ax2.grid(False)  # 去掉第二条线的网格
    ax2.plot(x, y2, label='平均播放量', color='#F75940', marker='o')
    ax1.set_yticks(np.arange(0, y1_max, y1_range)) #设置主坐标轴范围
    ax1.set_ylabel('发布视频数')  
    ax2.set_yticks(np.arange(0, y2_max, y2_range))   #设置次坐标轴范围
    ax2.set_ylabel('平均播放量')
    for tl in ax1.get_xticklabels():  # 旋转横轴标签
        tl.set_rotation(90)
    ax1.legend(loc=2)
    ax2.legend(loc=1)  # loc1代表放置右上方
    plt.title('2021年B站动画区视频发布时间及平均播放量分布')
    plt.show()
```

```python
plot_line_bar(hour_group, 60, 10,  200000, 50000)
```

```python
data2 = temp2[['UP主','发布时间','视频时长','点赞数','硬币数','收藏数','分享数','播放量','评论数','弹幕数','种类']]
AMV = data2[data2['种类'] == 'MAD']
MMV = data2[data2['种类'] == 'MMD']
sf = data2[data2['种类'] == '短片·手书·配音']   
gk = data2[data2['种类'] == '手办']
```

```python
print('AMV区共{}条数据\nMMV区共{}条数据\n短片-手书-配音区共{}条数据\n手办-模玩区共{}条数据'.format(len(AMV),len(MMV),len(sf),len(gk)))
```

```python
#I值计算
danmu_sum = AMV.groupby('UP主')['弹幕数'].sum()
comments_sum = AMV.groupby('UP主')['评论数'].sum()
views_sum = AMV.groupby('UP主')['播放量'].sum()
counts_sum = AMV.groupby('UP主')['发布时间'].count()

I = round((danmu_sum+comments_sum)/views_sum/counts_sum*100,2).reset_index() 
I.columns=['UP主','I']
I.head()
```

```python
#计算F值
earliest = AMV.groupby('UP主')['发布时间'].min()
latest = AMV.groupby('UP主')['发布时间'].max()
F = round((latest-earliest).dt.days/AMV.groupby('UP主')['发布时间'].count()).reset_index()
counts_sum = AMV.groupby('UP主')['发布时间'].count().reset_index()
counts_sum.columns = ['UP主','视频时长']
counts_sum1 = counts_sum[counts_sum['视频时长']<120]   #剔除发布视频数小于5的UP主

F = pd.merge(counts_sum1, F,on='UP主', how='inner')
F.columns = ['UP主','视频时长','F']
F.tail()
```

```python
#计算L值
AMV['likes_rate']=(AMV['点赞数']*1+AMV['硬币数']*2+AMV['点赞数']*3+AMV['分享数']*4)/AMV['播放量']*100
L = round((AMV.groupby('UP主')['likes_rate'].sum()/AMV.groupby('UP主')['发布时间'].count()).reset_index(),2)
L.columns = ['UP主', 'L']
L.head()
```

```python
#合并IFL值
IF = pd.merge(I,F,on = 'UP主',how = 'inner')
IFL = pd.merge(IF,L,on = 'UP主',how = 'inner')
IFL.head()
```
